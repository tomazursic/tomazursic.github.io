```python
# Function definitions

def get_alive(df):
    # Filter alive gateways
    alive = df[df.is_alive == True]
    print("Live gateways:", len(alive))
    return alive
    
def get_not_null_location(df):
    # Clean out null locations
    d = df[df.location.notnull()]
    print("Gateways with geo locations:", len(d))
    return d
```

## Get TTN Gateways data

```python
print(f"Today is: {datetime.now()}")

"""
NOTES:
This cell run slow!
Expect cca. >= 40582 rows

Data is stored in json file
"""

# Toggle this to force get latest data
# Default is one file per day
GET_FRESH_DATA = False

ROOT_PATH =  os.getcwd()
LOGS_DIR = os.path.join(os.getcwd(), "logs")

# Fetch data ttn v1
filename_v1 = today_filename()
f1 = os.path.join(LOGS_DIR, filename_v1)
print(f1)

# NOTE: Set from_file to False to fetch fresh data
# For new day is auto fetch new data
json_data_1 = get_data(
    f1, 
    APIURL, 
    from_file=os.path.isfile(f1),
    force=GET_FRESH_DATA
)

# Load all data into dataframe
df_ttnv1 = pd.DataFrame.from_dict(json_data_1)

# Transpose and set gw_id as index
df_ttnv1 = df_ttnv1.T

# Convert timestamps into datetime and add alive flags
df_ttnv1 = convert_timestamp(df_ttnv1, 'last_seen')

print("\n------\n")

# Fetch data TTN data 2
filename_v2 = today_filename(prefix=True)
f2 = os.path.join(LOGS_DIR, filename_v2)
print(f2)

json_data_2 = get_data(
    f2, 
    APIURL2, 
    from_file=os.path.isfile(f2),
    force=GET_FRESH_DATA
)

# Read json
# Load all data into dataframe
df_ttnv2 = pd.DataFrame.from_dict(json_data_2['statuses'])

# Transpose and set gw_id as index
df_ttnv2 = df_ttnv2.T

# Convert timestamps into datetime and add alive flags
df_ttnv2 = convert_timestamp(df_ttnv2, 'timestamp')

print()
print("v1 - All gateways:", len(df_ttnv1))
print("v2 - All gateways:", len(df_ttnv2))
```
